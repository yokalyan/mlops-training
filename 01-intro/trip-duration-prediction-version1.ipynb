```python
df = pd.read_parquet('https://d37ci6vzurychx.cloudfront.net/trip-data/green_tripdata_2025-01.parquet')
```

```python
df.lpep_pickup_datetime = pd.to_datetime(df.lpep_pickup_datetime)
df.lpep_dropoff_datetime = pd.to_datetime(df.lpep_dropoff_datetime)
#df.drop('trip_duration', axis=1, inplace=True)
df['trip_duration'] = df.lpep_dropoff_datetime - df.lpep_pickup_datetime
df
```

```python
#time_delta = df.trip_duration.iloc[0]
#time_delta = time_delta.total_seconds()/60
#time_delta
df['trip_duration'] = df['trip_duration'].apply(lambda time_delta: time_delta.total_seconds()/60)
df

```

```python
#Filtering with line type 2 looks to have problem. We will eliminate this
# df = df[df.trip_type == 2]
#df
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.distplot(df.trip_duration)
```

```python
df.trip_duration.describe(percentiles=[0.90, 0.95, 0.97, 0.98, 0.99])
```

```python
#((df.trip_duration >= 0) & (df.trip_duration <= 60)).mean()

# Close to 50% of the trips are less than 1 min, so we will remove those.

```

```python
df = df[(df.trip_duration >= 1) & (df.trip_duration <= 60)]
df
```

```python
categorical_features = ['PULocationID', 'DOLocationID']
numerical_features = ['trip_distance']
df[categorical_features] = df[categorical_features].astype('str')
df[categorical_features].dtypes
```

```python
df.dtypes
```

```python
from sklearn.feature_extraction import DictVectorizer
dv = DictVectorizer()
df[categorical_features + numerical_features].iloc[:10].to_dict(orient='records')
```

```python
training_dict = df[categorical_features + numerical_features].to_dict(orient='records')
training_dict
```

```python
X_train = dv.fit_transform(training_dict)
X_train
```

```python
dv.feature_names_
```

```python
y_train = df.trip_duration.values
y_train
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(X_train, y_train)
```

```python
y_pred = lr.predict(X_train)
```

```python
sns.distplot(y_pred, label='prediction')
sns.distplot(y_train, label='true value')
plt.legend()
```

```python
from sklearn.metrics import mean_squared_error
mean_squared_error(y_pred, y_train, squared=False)
```
